```python
#根据比例填充数据
def fill_with_prob(data,column):
    missing_count = data[column].isnull().sum()
    # 计算现有类别的分布比例
    value_counts = data[column].value_counts(normalize=True)
    categories = value_counts.index.tolist()
    probabilities = value_counts.values
    # 按照现有比例生成填充值
    fill_values = np.random.choice(categories, size=missing_count, p=probabilities)
    # 填充缺失值
    data.loc[data[column].isnull(), column] = fill_values
    return data
```

```python
#填充数据并且删除无用项
def repair_data(raw_data,test=False):
    data=raw_data.copy()
    
    #将int类型转换成float
    data['tbp_lv_symm_2axis_angle']=data['tbp_lv_symm_2axis_angle'].astype(float)
    
    #按比例填充未满项    
    for column in data.columns:
        if data[column].isnull().sum() > 0:  # 如果该列存在空值
            data = fill_with_prob(data,column)
    #tbp_lv_location分布太过失衡，只保留simple版本
    data = data.drop(columns=['tbp_lv_location'])
    return data
```

```python
#object项转换成独热编码
def trans_object(data):
    columns = ['sex','anatom_site_general','tbp_lv_location_simple','tbp_tile_type']
    # 初始化编码器
    encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
    
    # 对列进行编码
    for column in columns:
        encoded_array = encoder.fit_transform(data[[column]])
        encoded_df = pd.DataFrame(
            encoded_array,
            columns=[f"{column}_{cat}" for cat in encoder.categories_[0]],  # 手动生成列名
            index=data.index
        )
        # 合并结果
        data = pd.concat([data, encoded_df], axis=1)
        data.drop(columns=[column], inplace=True)
        
    if 'tbp_lv_location_simple_Unknown' in data.columns:
        data = data.drop(columns='tbp_lv_location_simple_Unknown')
    
    return data
```

```python
def train_data_trans(postive_data,negative_data,n=1000):
    #分别对数据修补
    new_p = repair_data(postive_data)
    new_n = repair_data(negative_data)
    #进行采样
    sampled_p = resample(
        new_p,
        replace=True,
        n_samples=n,  
        random_state=42)
    sampled_n = resample(
        new_n,
        replace=False,
        n_samples=3*n,  
        random_state=42)
    #数据拼接并将其填满
    combined_data = pd.concat([sampled_p, sampled_n], axis=0)
    combined_data = combined_data.sample(frac=1, random_state=42).reset_index(drop=True)
    combined_data=trans_object(combined_data)
    return combined_data
    
```

```python
def submit_data_trans(submit_data):
    data = submit_data.drop(columns = 'target')
    r_data = repair_data(data)
    f_data =  trans_object(r_data)
    return f_data
```

```python
#特征值归一化
def normalize_data(data):
    scaler = MinMaxScaler()
    img_id = data['id']
    if 'target' in data.columns:
        target = data['target']
        features = data.drop(columns = ['target','id'])
        features = pd.DataFrame(
            scaler.fit_transform(features),
            columns=features.columns,
            index=features.index
        )
        return features,target
    else:
        features = data.drop(columns = ['id'])
        features = pd.DataFrame(
            scaler.fit_transform(features),
            columns=features.columns,
            index=features.index
        )
        return features
```

```python
#图像数据
class ImageDataset(Dataset):
    def __init__(self, data, image_dir, transform=None):
        self.labels = data['target'].values if 'target' in data.columns else None
        self.image_ids = data['id'].values
        self.image_dir = image_dir
        self.transform = transform

    def __len__(self):
        return len(self.image_ids)

    def __getitem__(self, idx):
        img_id = self.image_ids[idx]
        label = self.labels[idx] if self.labels is not None else 'UNK'

        # 构造图像路径
        img_path = os.path.join(self.image_dir, f"{img_id}.jpg")

        
        # 加载图像
        image = Image.open(img_path).convert('RGB')

        # 应用图像预处理
        if self.transform:
            image = self.transform(image)

     

        return image, label


        
```

```python
#图像格式修改
def get_resnet_transform():
    transform = transforms.Compose([
        transforms.Resize((224, 224)),  
        transforms.ToTensor(),         
        transforms.Normalize(
            mean=[0.485, 0.456, 0.406],  
            std=[0.229, 0.224, 0.225]
        )
    ])
    return transform

def imgdataloader(data,image_dir, batch_size=32, shuffle=True):
    #创建dataloader训练图像模型
    dataset = ImageDataset(
        data=data,
        image_dir=image_dir,
        transform=get_resnet_transform()
    )
    dataloader = torch.utils.data.DataLoader(
        dataset,
        batch_size=batch_size,
        shuffle=shuffle
    )
    return dataloader
    
```

```python
#resnet18 作为图模型
def create_resnet18(num_classes=2):
    model = models.resnet18(pretrained=True)
    model.fc = nn.Linear(model.fc.in_features, num_classes)  # 修改全连接层
    return model
```

```python
#构建对于文本数据的模型
class MLP(nn.Module):
    def __init__(self, input_size):
        super(MLP, self).__init__()
        self.fc1 = nn.Linear(input_size, 128)
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, 32)
        self.fc4 = nn.Linear(32,1)
        self.dropout = nn.Dropout(0.5)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = self.dropout(x)
        x = torch.relu(self.fc2(x))
        x = self.dropout(x)
        x = torch.relu(self.fc3(x))
        x = torch.sigmoid(self.fc4(x))
        
        return x
```

```python
class MLP_f(nn.Module):
    def __init__(self, input_size):
        super(MLP_f, self).__init__()
        self.fc1 = nn.Linear(input_size, 128)
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, 32)
        self.fc4 = nn.Linear(32,16)
        self.fc5 = nn.Linear(16,1)
        self.dropout = nn.Dropout(0.1)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = self.dropout(x)
        x = torch.relu(self.fc2(x))
        x = self.dropout(x)
        x = torch.relu(self.fc3(x))
        x = self.dropout(x)        
        x = torch.relu(self.fc4(x))
        x = torch.sigmoid(self.fc5(x))
        
        return x
```

```python
def train_one_epoch(model, dataloader, criterion, optimizer, device):
    
    model.train()
    total_loss = 0
    for images, labels in dataloader:
        images, labels = images.to(device), labels.to(device)


        # 前向传播
        outputs = model(images)
        loss = criterion(outputs, labels)

        # 反向传播
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
    return total_loss / len(dataloader)


def validate_img_collect_roc(model, dataloader, criterion, device):
    model.eval()
    total_loss = 0
    correct = 0
    total = 0
    probs = []
    labels = []

    with torch.no_grad():
        for images, targets in dataloader:
            images, targets = images.to(device), targets.to(device)

            outputs = model(images)
            loss = criterion(outputs, targets)
            total_loss += loss.item()

            # 获取预测概率
            probabilities = torch.softmax(outputs, dim=1)[:, 1]  # 取正类的概率
            preds = probabilities > 0.5  # 假设阈值为0.5

            probs.extend(probabilities.cpu().numpy())
            labels.extend(targets.cpu().numpy())

            correct += (preds == targets).sum().item()
            total += targets.size(0)

    accuracy = correct / total
    return total_loss / len(dataloader), accuracy, probs, labels


def validate_tabular_collect_roc(model, dataloader, criterion, device):
    model.eval()
    total_loss = 0
    correct = 0
    total = 0
    probs = []
    labels = []

    with torch.no_grad():
        for features, targets in dataloader:
            features, targets = features.to(device), targets.to(device)

            outputs = model(features)
            loss = criterion(outputs, targets)
            total_loss += loss.item()

            # 获取预测概率
            probabilities = outputs.squeeze(1)  # 二分类模型的输出是一个值
            preds = (outputs > 0.5).float()  # 假设阈值为0.5

            probs.extend(probabilities.cpu().numpy())
            labels.extend(targets.cpu().numpy())

            correct += (preds == targets).sum().item()
            total += targets.size(0)

    accuracy = correct / total
    return total_loss / len(dataloader), accuracy, probs, labels

```

```python
from torch.optim import Adam

def create_folds(data, k):
    skf = StratifiedKFold(n_splits=k, shuffle=True, random_state=42)
    labels = data['target'].values
    folds = list(skf.split(data, labels))
    return folds


def train_img_kfold(data, image_dir, num_epochs=10, batch_size=32, lr=0.001, k=5):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    folds = create_folds(data, k)
    results = []
    best_model_paths = []
    all_probs = []  # 用于存储所有折的验证集预测概率
    all_labels = []  # 用于存储所有折的验证集真实标签

    for fold, (train_idx, val_idx) in enumerate(folds):
        print(f"Training Fold {fold + 1}/{k}")

        # 创建训练集和验证集的 DataLoader
        train_data = data.iloc[train_idx]
        val_data = data.iloc[val_idx]
        train_loader = imgdataloader(train_data, image_dir, batch_size=batch_size, shuffle=True)
        val_loader = imgdataloader(val_data, image_dir, batch_size=batch_size, shuffle=False)

        # 创建模型、损失函数和优化器
        model = create_resnet18(num_classes=2).to(device)
        criterion = nn.CrossEntropyLoss()
        optimizer = Adam(model.parameters(), lr=lr)

        # 初始化最佳验证准确率
        best_val_accuracy = 0.0
        best_model_path = f"best_img_model_fold_{fold + 1}.pth"

        # 训练与验证
        for epoch in range(num_epochs):
            train_loss = train_one_epoch(model, train_loader, criterion, optimizer, device)
            val_loss, val_accuracy, fold_probs, fold_labels = validate_img_collect_roc(
                model, val_loader, criterion, device
            )
            print(f"Epoch {epoch + 1}: Train Loss = {train_loss:.4f}, Val Loss = {val_loss:.4f}, Val Accuracy = {val_accuracy:.4f}")

            # 保存验证准确率最好的模型
            if val_accuracy > best_val_accuracy:
                best_val_accuracy = val_accuracy
                torch.save(model.state_dict(), best_model_path)

        # 保存当前折的验证结果和模型路径
        results.append(best_val_accuracy)
        best_model_paths.append(best_model_path)
        all_probs.extend(fold_probs)
        all_labels.extend(fold_labels)

        print(f"Best Model for Fold {fold + 1} saved at {best_model_path} with Val Accuracy: {best_val_accuracy:.4f}")

    # 打印总体结果
    print(f"Average Validation Accuracy: {sum(results) / len(results):.4f}")
    return results, best_model_paths, all_probs, all_labels


```

```python
def train_tabular_kfold(model_name, model_class, data, labels, input_size, num_epochs, batch_size, lr, k):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    
    # 确保输入是 NumPy 格式
    data = data.to_numpy() if isinstance(data, pd.DataFrame) else data
    labels = labels.to_numpy() if isinstance(labels, (pd.DataFrame, pd.Series)) else labels

    # 创建 TensorDataset
    dataset = TensorDataset(torch.tensor(data, dtype=torch.float32), 
                            torch.tensor(labels, dtype=torch.float32).unsqueeze(1))
    kf = KFold(n_splits=k, shuffle=True, random_state=42)
    
    results = []
    best_model_paths = []
    all_probs = []  # 用于存储所有折的验证集预测概率
    all_labels = []  # 用于存储所有折的验证集真实标签

    for fold, (train_idx, val_idx) in enumerate(kf.split(dataset)):
        print(f"Training Fold {fold + 1}/{k}")
        
        # 创建训练和验证集的 DataLoader
        train_subset = Subset(dataset, train_idx)
        val_subset = Subset(dataset, val_idx)
        train_loader = DataLoader(train_subset, batch_size=batch_size, shuffle=True)
        val_loader = DataLoader(val_subset, batch_size=batch_size, shuffle=False)

        # 初始化模型、损失函数和优化器
        model = model_class(input_size=input_size).to(device)
        criterion = torch.nn.BCELoss()  # 二分类交叉熵损失
        optimizer = torch.optim.Adam(model.parameters(), lr=lr)

        best_accuracy = 0.0
        best_model_path = f"best_{model_name}_model_fold_{fold+1}.pth"
        
        # 训练和验证
        for epoch in range(num_epochs):
            train_loss = train_one_epoch(model, train_loader, criterion, optimizer, device)
            val_loss, val_accuracy, fold_probs, fold_labels = validate_tabular_collect_roc(
                model, val_loader, criterion, device
            )
            print(f"Epoch {epoch+1}: Train Loss = {train_loss:.4f}, Val Loss = {val_loss:.4f}, Val Accuracy = {val_accuracy:.4f}")
            
            # 保存最优模型
            if val_accuracy > best_accuracy:
                best_accuracy = val_accuracy
                torch.save(model.state_dict(), best_model_path)

        # 保存当前折的验证结果和模型路径
        results.append(best_accuracy)
        best_model_paths.append(best_model_path)
        all_probs.extend(fold_probs)
        all_labels.extend(fold_labels)
        print(f"Best Model for Fold {fold + 1} saved at {best_model_path} with Val Accuracy: {best_accuracy:.4f}")

    # 打印总体结果
    print(f"Average Validation Accuracy: {np.mean(results):.4f}")
    return results, best_model_paths, all_probs, all_labels
```

```python
#生成基于图像的平均概率
def ensemble_imgpredict_features(model_class, num_folds, test_loader, device):
    all_predictions = []

    for fold in range(1, num_folds + 1):
        # 加载当前折的模型
        model = model_class(num_classes=2).to(device)
        model.load_state_dict(torch.load(f"best_img_model_fold_{fold}.pth"))
        model.eval()

        fold_predictions = []
        with torch.no_grad():
            for batch in test_loader:
                images, _ = batch  # 始终有 `_` 作为占位符
                images = images.to(device)
                outputs = model(images)
                probs = torch.softmax(outputs, dim=1)[:, 1]  # 取正类概率
                fold_predictions.extend(probs.cpu().tolist())

        all_predictions.append(fold_predictions)

    final_predictions = np.mean(all_predictions, axis=0)
    return final_predictions

```

```python
def plot_roc_curve(probs, labels):
    fpr, tpr, thresholds = roc_curve(labels, probs)
    roc_auc = auc(fpr, tpr)

    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f"ROC Curve (AUC = {roc_auc:.2f})")
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend(loc="lower right")
    plt.show()

```
